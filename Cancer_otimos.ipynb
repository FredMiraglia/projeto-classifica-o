```python
X = df.drop('Alvo', axis=1)
y = df['Alvo'].replace([0,1],['benigno','maligno'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
X_smote = smote.fit_resample(X,y)
```

```python
X_train, X_test, y_train,y_test =train_test_split(X_smote[0],X_smote[1], test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
print(confusion_matrix(y_test, pred))
```

```python
pred_total = model.predict(X)
```

```python
print(classification_report(y, pred_total))
print(confusion_matrix(y, pred_total))
```
